## matplotlib的学习和研究


### 用官方的一个例子引入学习的过程
官方给出了一个双柱状图的案例, 同时, 增加了数据标签

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

x = np.arange(len(labels))  # the label locations
width = 0.4  # the width of the bars

fig, ax = plt.subplots() #设置了一个图版ax
fig, ay = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='Men')
rects2 = ax.bar(x + width/2, women_means, width, label='Women')

info = ay.bar(x, men_means, width, label = 'Men')
ay.set_xticks(x)
ay.set_xticklabels(labels)
ay.legend()


# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()  #获取高度
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```

```python
for rect in rects1:
  print('%s的高度是%s'%(rect,rect.get_height()))
```

可以看到, 一个绘制图的参数由以下内容组成:

`xy` 定义了这个图形的位置;

`width`定义了bar的宽度, 对于其他图形, 由不同的参数

`height`定义了bar的高度

`angle`定义了bar的角度

在`pylab`的状态下, 可以直接调用plt下的命令, 所以不需要`plt.`


```python
# 创建一个 8 * 6 点（point）的图，并设置分辨率为 80
plt.figure(figsize=(8,4.5), dpi=80)
# 创建一个新的 1 * 1 的子图，接下来的图样绘制在其中的第 1 块（也是唯一的一块）
plt.subplot(1,1,1)
X = np.linspace(-np.pi, np.pi, 256,endpoint=True)
C,S = np.cos(X), np.sin(X)
# 绘制余弦曲线，使用蓝色的、连续的、宽度为 1 （像素）的线条
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-")
# 绘制正弦曲线，使用绿色的、连续的、宽度为 1 （像素）的线条
plt.plot(X, S, color="red", linewidth=2.5, linestyle="-")
# 设置横轴记号
#plt.xticks(np.linspace(-4,4,9,endpoint=True))
# 设置纵轴记号
#plt.yticks(np.linspace(-1,1,5,endpoint=True))
#下面更新了坐标轴
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],[r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, +1],[r'$-1$', r'$0$', r'$+1$'])
# 设置横轴的上下限
#plt.xlim(-4.0,4.0)
# 设置纵轴的上下限
#plt.ylim(-1.0,1.0)
# 下面更新了x,y轴的长度, 使得两端外延了20%, 优化显示面积
xmin ,xmax = X.min(), X.max()
ymin, ymax = C.min(), C.max()
dx = (xmax - xmin) * 0.1
dy = (ymax - ymin) * 0.1
plt.xlim(xmin - dx, xmax + dx)
plt.ylim(ymin - dy, ymax + dy)

# 以分辨率 72 来保存图片
# savefig("exercice_2.png",dpi=72)
# 在屏幕上显示
plt.show()
```

下面移动了图形的坐标轴


```python
# 创建一个 8 * 6 点（point）的图，并设置分辨率为 80
plt.figure(figsize=(8,4.5), dpi=80)
# 创建一个新的 1 * 1 的子图，接下来的图样绘制在其中的第 1 块（也是唯一的一块）
plt.subplot(1,1,1)
X = np.linspace(-np.pi, np.pi, 256,endpoint=True)
C,S = np.cos(X), np.sin(X)
# 绘制余弦曲线，使用蓝色的、连续的、宽度为 1 （像素）的线条
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-", label='cos')
# 绘制正弦曲线，使用绿色的、连续的、宽度为 1 （像素）的线条
plt.plot(X, S, color="red", linewidth=2.5, linestyle="-", label= 'sin')
# 设置横轴记号
#plt.xticks(np.linspace(-4,4,9,endpoint=True))
# 设置纵轴记号
#plt.yticks(np.linspace(-1,1,5,endpoint=True))
#下面更新了坐标轴
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],[r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, +1],[r'$-1$', r'$0$', r'$+1$'])
# 设置横轴的上下限
#plt.xlim(-4.0,4.0)
# 设置纵轴的上下限
#plt.ylim(-1.0,1.0)
# 下面更新了x,y轴的长度, 使得两端外延了20%, 优化显示面积
xmin ,xmax = X.min(), X.max()
ymin, ymax = C.min(), C.max()
dx = (xmax - xmin) * 0.1
dy = (ymax - ymin) * 0.1
plt.xlim(xmin - dx, xmax + dx)
plt.ylim(ymin - dy, ymax + dy)

# 移动坐标轴的方式
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

plt.legend(loc='upper left')
# 以分辨率 72 来保存图片
# savefig("exercice_2.png",dpi=72)
# 在屏幕上显示

t = 2*np.pi/3
#画了线和点
plt.plot([t,t],[0,np.cos(t)], color ='blue', linewidth=2.5, linestyle="--")
plt.scatter([t,],[np.cos(t),], 50, color ='blue')
#画一个标注信息
#plt.annotate(r'$\sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$',
#         xy=(t, np.sin(t)), xycoords='data',
#         xytext=(+10, +30), textcoords='offset points', fontsize=16,
#         arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

#plot([t,t],[0,np.sin(t)], color ='red', linewidth=2.5, linestyle="--")
#scatter([t,],[np.sin(t),], 50, color ='red')

plt.annotate(r'$\cos(\frac{2\pi}{3})=-\frac{1}{2}$',
         xy=(t, np.cos(t)), xycoords='data',
         xytext=(-90, -50), textcoords='offset points', fontsize=16,
         arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))


plt.show()
```

# FAA鸟撞数据
数据来源为FAA公布的Wild Strikes Databas, 地址为: [wildlife](https://wildlife.faa.gov/home)

最新的数据版本为Access, 此处采用的是-15年的数据, 来源为kaggle

```python
database = './database.csv'
data = pd.read_csv(database, low_memory=False)
```

当前数据总量包括:

```python
data.iloc[:,0].count()
```

```python
data.columns
```

```python
import seaborn as sns
```

```python
sby = data.groupby('Incident Year')
```

```python
strike = ['Radome Strike', 'Windshield Strike', 'Nose Strike', 'Engine1 Strike', 'Engine2 Strike', 'Engine3 Strike',
          'Engine4 Strike', 'Propeller Strike', 'Wing or Rotor Strike', 'Fuselage Strike', 'Landing Gear Strike',
          'Tail Strike', 'Lights Strike', 'Other Strike']
```

```python
sby_sum = pd.DataFrame(sby['Incident Year'].count())
```

```python
sby_sum.head()
```

```python
sby_sum.columns = ['counts']
sby_sum = sby_sum.reset_index()
```

```python
sby_sum.iterrows()
```

```python
for i,r in sby_sum.iterrows():
    print ('%s,%s\n'%(i,r.counts))
```

## 时间变化

```python
plt.figure(figsize=(8,4.5), dpi=80)
info = sns.barplot(x=(sby_sum['Incident Year']),y=(sby_sum['counts']),data = sby_sum)
plt.xticks(rotation=90)
for i,r in sby_sum.iterrows():
    plt.text(i,(r['counts']+10),r['counts'],color="black",ha="center")
plt.show()
```

```python
sbph = data.groupby('Flight Phase')
```

```python
sbph[strike].sum()
```

```python
sbph_s = sbph[strike].sum()
sbph_s.iloc[:,0].count()
```

```python
sbph_s.iloc[0,:].count()
```

```python
sum = []
for i in range(12):
    sum.append(sbph_s.iloc[i,:].sum())
sbph_s['sum']=sum
```

### 评估阶段和撞击位置


```python
sbph_s
```

```python
plt.figure(figsize=(8,4.5), dpi=80)
sns.barplot(sbph_s.index,sbph_s['sum'],data = sbph_s)
plt.xticks(rotation=45)
show()
```
